### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing the dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
x2 = iris.data[:,2:]
x1 = iris.data[:,:2]
y = iris.target
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X1 = StandardScaler()
X1 = sc_X1.fit_transform(X1)
sc_X2 = StandardScaler()
X2 = sc_X2.fit_transform(X2)
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(x1, y, test_size = 0.2, random_state = 42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(x2, y, test_size = 0.2, random_state = 42)
```

### Fitting SVR to the dataset with RBF

```python
from sklearn.svm import SVC
Model_linear_x1 = SVC(kernel = 'linear', C=100, random_state = 42 )
Model_linear_x1.fit(X_train1, y_train1)
y_pred1 = Model_linear_x1.predict(X_test1)

rounded_y_pred1=[]
for num in y_pred1:
    rounded_num = round(num)  # Round each number to 2 decimal places
    rounded_y_pred1.append(rounded_num)
y_pred1=  np.array(rounded_y_pred1)
```

###score

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score,f1_score
cm = confusion_matrix(y_test1, y_pred1)
print("confusion matrix :")
print(cm)
ac =accuracy_score(y_test1, y_pred1)
ps=precision_score(y_test1, y_pred1 ,average='macro')
rs=recall_score(y_test1, y_pred1 ,average='macro')
f1=f1_score(y_test1, y_pred1 ,average='macro')
print(' accuracy_score =  %.3f    precision_score =  %.3f   recall_score =   %.3f   f1_score =   %.3f '
       %(ac,ps,rs,f1))
```

### Visualising the SVR results test

```python
from matplotlib.colors import ListedColormap
X_set, y_set = X_test1, y_test1
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, Model_linear_x1.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.5, cmap = ListedColormap(('red', 'blue','green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
kind = ["Iris-setosa","Iris-versicolor","Iris-virginica"]
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'blue','green'))(i), label = kind[j])
plt.title('SVM test')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.legend()
plt.show()
```
