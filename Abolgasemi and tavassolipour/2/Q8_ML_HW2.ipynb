## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

## Importing the dataset

```
dataset = pd.read_csv('ted_main.csv')
X = dataset.iloc[ : , 2].values.reshape(-1, 1)
bandwidths = [1, 10, 20, 50, 100]
plt_ind = np.arange(6) + 231
x = np.linspace(0, 7000, 100000)


for bandwidth, ind in zip(bandwidths, plt_ind):
  pdf = [0]*100000
  for i in range(0, len(X)):
     pdf = norm.pdf(x, X[i], bandwidth) + pdf
  plt.plot(x, pdf)
  plt.title("h="+str(bandwidth))
  plt.show()
```

```
bandwidths = [1, 10, 20, 30, 50, 100]
x = np.linspace(0, 7000, 100000)

from sklearn.neighbors import KernelDensity
for i in bandwidths:
  for j in range(250,2000,250):
    kde = KernelDensity(bandwidth=i, kernel='gaussian')
    choosed_data = X[:j]
    kde.fit(choosed_data)
    log_dens = kde.score_samples(x.reshape(-1, 1))
    f = np.exp(log_dens)
    plt.plot(x, f,label = "n = %d" % j)
    plt.legend()
  plt.title("h="+ str(i))
  integral = np.trapz(f, x)
  plt.show()
```
