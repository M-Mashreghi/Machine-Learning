```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

loc_a1 = 3
loc_a2 = 5
scale_b = 1
prior_w2=0.5;
prior_w1=0.5;
middle = (loc_a1+loc_a2)/2
x = np.linspace(-6, 14, 100000)  
Cauchy1 = (1/((np.pi)*scale_b))*(1/(1+(((x-loc_a1)/scale_b)**2)));
Cauchy2 = (1/((np.pi)*scale_b))*(1/(1+(((x-loc_a2)/scale_b)**2)));

plt.axvline(alpha=0.8,x = middle, lw=1, color = 'pink', label = 'decision boundary')
plt.plot(x, Cauchy1, 'r-', lw=2, alpha=0.6,label="$P(x|w_{1}$)")
plt.plot(x, Cauchy2, 'b-', lw=2, alpha=0.6,label="$P(x|w_{2}$)")
plt.legend()
plt.show()
#now caculating next p(w1|x) &  p(w2|x)
px = (Cauchy1*prior_w1) + (Cauchy2*prior_w2);
Posterior_w1 = (Cauchy1*prior_w1)/px;
Posterior_w2 = (Cauchy2*prior_w2)/px;

plt.axvline(alpha=0.8,x = middle, lw=1, color = 'pink', label = 'decision boundary')
plt.plot(x, Posterior_w1, 'r-', lw=2, alpha=0.6,label="$P(w_{1}$|x)")
plt.plot(x, Posterior_w2, 'b-', lw=2, alpha=0.6,label="$P(w_{2}$|x)")
plt.legend()
plt.show()
```
