```python
import numpy as np
import matplotlib.pyplot as plt

X1_red = np.array([0,1.5,2,1,2.5,1.5,0.5,1,2])
Y1_red = np.array([0.5,0,1,1,2,2,2,3,3])

X2_black = np.array([-2,-1.5,-1,-1.5,-0.5,0.5,0.5,1,1.5,1.5])
Y2_black = np.array([-1,0,-1,1,0.5,0.5,-0.5,-1,-0.5,0.5])


x = np.linspace(-8, 8, 300)
y = np.linspace(-8, 8, 300)

X, Y = np.meshgrid(x, y)
```

##A

```python


def f_A(x, y):
    return 5.51*x**2 - 3.36*x*y - 21*x - 5*y**2 - 9.2*y + 21.45
Z_A = f_A(X, Y)

plt.contour(X, Y, Z_A, levels=[0], colors='red')

plt.scatter(X1_red, Y1_red, color='red')

plt.scatter(X2_black, Y2_black, color='black')

plt.xlim(-3, 3)
plt.ylim(-2, 4)

plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title("PART A")
plt.legend(['Red Points', 'Black Points'])
plt.show()
```
